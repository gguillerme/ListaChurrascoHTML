```python
from google.colab import drive

# Montar o Google Drive
drive.mount('/content/drive')

# Função para calcular a quantidade de comida e bebida necessária para o churrasco
def calcular_consumo(desejos, num_convidados):
    consumo_total = {}

    # Calculando a quantidade total de cada item desejado
    for item, quantidade in desejos.items():
        consumo_total[item] = quantidade * num_convidados

    return consumo_total

# Função para gerar a página HTML com os resultados
def gerar_pagina_html(nome_aluno, desejos, consumo_total):
    html = "<!DOCTYPE html>\n<html>\n<head>\n<title>CHURRASCO - " + nome_aluno + "</title>\n</head>\n<body>\n"
    html += "<h1>CHURRASCO - " + nome_aluno + "</h1>\n"
    html += "<h2>Lista de produtos e quantidades para comprar:</h2>\n"
    html += "<ul>\n"

    for item, quantidade in consumo_total.items():
        html += "<li>" + item + ": " + str(quantidade) + "</li>\n"

    html += "</ul>\n</body>\n</html>"

    # Salvando o HTML no Google Drive
    with open("/content/drive/My Drive/churrasco_" + nome_aluno + ".html", "w") as f:
        f.write(html)

# Função principal
def main():
    nome_aluno = input("Digite seu nome: ")

    # Recebendo lista de desejos
    desejos = {}
    num_itens = int(input("Quantos itens você deseja incluir na lista? (Entre 10 e 20): "))
    while num_itens < 10 or num_itens > 20:
        print("Por favor, escolha um número entre 10 e 20.")
        num_itens = int(input("Quantos itens você deseja incluir na lista? (Entre 10 e 20): "))

    for i in range(num_itens):
        item = input("Digite o item desejado: ")
        quantidade = int(input("Digite a quantidade desejada para " + item + ": "))
        desejos[item] = quantidade

    # Definindo o número de convidados
    num_convidados = 30

    # Calculando o consumo total
    consumo_total = calcular_consumo(desejos, num_convidados)

    # Gerando a página HTML com os resultados
    gerar_pagina_html(nome_aluno, desejos, consumo_total)

    print("Página HTML gerada com sucesso e salva no Google Drive!")

# Chamando a função principal
if __name__ == "__main__":
    main()
```
